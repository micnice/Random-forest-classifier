```python
df = df[df.value!='REDEPLOYE']
df = df[df.value!='O']
df = df[df.value!='Discordance imbécile']
df = df[df.value!='rupture']
df = df[df.value!='Non Applicable (N.A)']
df = df[df.value!='ssc']
df = df[df.value!='périmés']
df = df[df.value!='flaçon de 10doses']
df = df[df.value!='50000 au BCZS-Moba']
df = df[df.value!='Expiré']
df = df[df.value!='flaçon 10doses']
df = df[df.value!='Quantité à risque de péremption']
df = df[df.value!='EXPIRE']
df = df[df.value!='10 doses']
df = df[df.value!='Rupture']
df = df[df.value!='péremption']
df = df[df.value!='expiré']
df = df[df.value!='sont périmés']
df = df[df.value!='ADBC']
df = df[df.value!='10 VERS BON BERGE']
df = df[df.value!='300mg']
df = df[df.value!='les activités PEV ne sont pas intégré']
df = df[df.value!='30R']
df = df[df.value!='NaN']
```

```python
# drop columns with less than 1000 values
df = df.drop(['8/1/2018','9/1/2018','8/1/2016','6/1/2018','12/1/2018','11/1/2018','11/1/2016','10/1/2018','10/1/2016'], axis=1)
df
```

```python
#find roles with just one numeric value
df['row_values'] = df.count(1)
df
```

```python
#Drop  rows with only one numeric value
df = df[df['row_values'] > 3]
df = df.drop(['row_values'], axis=1)
df
```

```python
# Find rows with  all zero values
df['row_total'] = df.sum(axis=1)
df
```

```python
# drop roles with a max of 0
df = df[df['row_total'] != 0]
df = df.drop(['row_total'], axis=1)
df
```

```python
#replace NAN with row mean
m = df.mean(axis=1)
for i, col in enumerate(df):
    # using i allows for duplicate columns
    # inplace *may* not always work here, so IMO the next line is preferred
    # df.iloc[:, i].fillna(m, inplace=True)
    df.iloc[:, i] = df.iloc[:, i].fillna(m)
df
```
